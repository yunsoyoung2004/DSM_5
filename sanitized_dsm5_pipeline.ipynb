Synthetic placeholder data is used to demonstrate the full experimental pipeline and labeling logic.

```
# =========================================================
# Public Release Version (Sanitized)
# - Raw data paths removed
# - No real text displayed or stored
# - Uses synthetic placeholder data
# =========================================================

import pandas as pd
import numpy as np

try:
    from konlpy.tag import Kkma
    kkma = Kkma()
except:
    kkma = None

np.random.seed(42)
NUM_SAMPLES = 1000

utter = pd.DataFrame({"text": ["SYNTHETIC_TEXT"] * NUM_SAMPLES})

utter_kkma = []
for i in range(len(utter)):
    if kkma is not None:
        utter_kkma.append(kkma.pos(utter['text'][i]))
    else:
        utter_kkma.append([('SYN', 'NNG')])

rewords = {
    1: [['우울'], ['슬프']],
    2: [['흥미', '없']],
    3: [['식욕']],
    4: [['잠']],
    5: [['초조']],
    6: [['피로']],
    7: [['자책']],
    8: [['집중']],
    9: [['자살']]
}

results = []
for i in range(len(utter_kkma)):
    tokens = [x for (x, y) in utter_kkma[i]]
    for label, rules in rewords.items():
        for rule in rules:
            if all(k in tokens for k in rule):
                results.append({'index': i, 'label': label})

results_df = pd.DataFrame(results)
label_counts = results_df['label'].value_counts().sort_index()
label_counts
```
